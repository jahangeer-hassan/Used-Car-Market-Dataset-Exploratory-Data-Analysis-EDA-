# Used Car Market Dataset Exploratory Data Analysis (EDA)

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Introduction

This project performs Exploratory Data Analysis (EDA) on a used car market dataset.  
The dataset contains information about car brands, fuel types, prices, mileage, transmission types, and cities.

```
df = pd.read_csv("/content/final.csv")
```

## Dataset Overview

The dataset helps analyze different factors affecting used car prices and market trends.

##  What are the first five rows of the dataset?

```
df.head()
```

## What are the last five rows of the dataset?

```
df.tail()
```

## What is the shape of the dataset?

```
df.shape
```

## What is the basic information of the dataset?

```
df.info()
```

## What is the statistical summary of the dataset?

```
df.describe()
```

## Are there any missing values in the dataset?

```
df.isnull().sum()
```

##  Car Brand Analysis

##  Which car brand is most common in the dataset?

```
df["brand"].value_counts()
```

## Top Car Brands Bar Chart

```
import matplotlib.pyplot as plt

df["brand"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Top Car Brands")
plt.show()
```

## Fuel Type Analysis

## What is the distribution of fuel types?

```
df["fuel_type"].value_counts()
```

## Fuel Type Distribution Pie Chart

```
df["fuel_type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Fuel Type Distribution")
plt.show()
```

## Price Type Analysis

##  What is the average listed price of cars?

```
df["listed_price"].mean()
```

##  What are the maximum and minimum car prices?

```
print(df["listed_price"].max())
print(df["listed_price"].min())
```

## Car Price Distribution Histogram

```
df["listed_price"].hist(bins=20)
plt.xlabel("Listed Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

## Car Age Analysis

```
df["car_age"].mean()
```

## Car Age Distribution Histogram

```
df["car_age"].hist(bins=15)
plt.xlabel("Car Age")
plt.ylabel("Count")
plt.title("Car Age Distribution")
plt.show()
```

## Mileage Analysis

##  What is the average mileage of cars?

```
df["mileage"].mean()
```

## Mileage vs Listed Price Scatter Plot

```
plt.scatter(df["mileage"], df["listed_price"])
plt.xlabel("Mileage")
plt.ylabel("Listed Price")
plt.title("Mileage vs Price")
plt.show()
```

## Transmission Analysis

##  Which transmission type is more common?

```
df["transmission"].value_counts()
```

## Transmission Type Plot Chart

```
df["transmission"].value_counts().plot(kind="bar")
plt.xlabel("Transmission")
plt.ylabel("Count")
plt.title("Transmission Type")
plt.show()
```

## City Analysis

## Which city has the highest number of listed cars?

```
df["city"].value_counts()
```

## Cars by City Plot Chart

```
df["city"].value_counts().plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Count")
plt.title("Cars by City")
plt.show()
```

## Relationship Analysis

## Is there any relationship between car age and listed price?

## Car Age vs Listed Price Scatter Plot

```
plt.scatter(df["car_age"], df["listed_price"])
plt.xlabel("Car Age")
plt.ylabel("Listed Price")
plt.title("Car Age vs Price")
plt.show()
```

## What is the relationship between kilometers driven and car price?

## KM Driven vs Price Scatter Plot

```
plt.scatter(df["km_driven"], df["listed_price"])
plt.xlabel("KM Driven")
plt.ylabel("Listed Price")
plt.title("KM Driven vs Price")
plt.show()
```

## Correlation Heatmap

```
import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Final Conclusion
This Exploratory Data Analysis (EDA) project analyzed the used car market dataset to understand the relationship between variables like car price, fuel type, transmission, mileage, and kilometers driven. The analysis showed that newer cars with lower mileage usually have higher prices, while older cars with more kilometers driven have lower resale value. Different graphs and visualizations helped identify market trends, customer preferences, and pricing patterns in the used car industry.
